```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random as rand
```

```python
# The formula for the logistic model: r*xn*(1-xn)
def logistic(xn,r):
    return r*xn - r*xn*xn;
```

```python
def getPoints(x0,r,nsettle,nrec):
    xn = x0
    xv = np.zeros(nrec)
    for i in range(nsettle):
        xn = logistic(xn,r)
    for j in range(nrec):
        xn = logistic(xn,r)
        xv[j] = xn
    return xv
```

```python
x0 = rand.random()
# number of iterations before you start recording values
# this allows the map to settle
nsettle = 200
# number of iterations recorded for each value of r
nrec = 500
# number of values of r taken between 1 and 4
nr = 300
rv = np.linspace(1.0+3.0/nr,4.0,num = nr)
rr = np.empty(nrec)
rr.fill(1.0)
xx = getPoints(x0,1.0,nsettle,nrec)
for r in rv:
    rr1 = np.empty(nrec)
    rr1.fill(r)
    rr = np.concatenate((rr,rr1))
    xx1 = getPoints(x0,r,nsettle,nrec)
    xx = np.concatenate((xx,xx1))
plt.figure(figsize=(20,10))
plt.xlabel("r",fontsize="xx-large")
plt.ylabel("x",fontsize="xx-large")
plt.plot(rr,xx,'.')
plt.show()
```
